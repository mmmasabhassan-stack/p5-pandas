```python
import random
import csv

# Pool of 100 Pakistani boy names
pakistani_boy_names = [
    "Ahmed", "Ali", "Hassan", "Hussain", "Hamza", "Bilal", "Usman", "Omar", "Yasir",
    "Imran", "Zubair", "Farhan", "Faisal", "Adeel", "Tariq", "Asad", "Shahbaz", "Junaid",
    "Kamran", "Noman", "Salman", "Ahsan", "Waqar", "Saad", "Zeeshan", "Adnan", "Amir",
    "Rashid", "Sajid", "Shahid", "Talha", "Shafiq", "Muneeb", "Rizwan", "Arif", "Naveed",
    "Majid", "Kashif", "Raheel", "Shahzad", "Furqan", "Qasim", "Danish", "Arslan", "Aqeel",
    "Taha", "Ammar", "Mudassir", "Azhar", "Sufyan", "Irfan", "Nouman", "Atif", "Sarfaraz",
    "Mubashir", "Rauf", "Sami", "Obaid", "Farrukh", "Faizan", "Sabir", "Jawad", "Moiz",
    "Younus", "Ilyas", "Haroon", "Jibran", "Tanveer", "Akram", "Aslam", "Munir", "Rehan",
    "Imtiaz", "Shafi", "Arham", "Mustafa", "Zain", "Hamad", "Rameez", "Ayaz", "Abbas",
    "Haider", "Shuja", "Saif", "Waleed", "Anwar", "Bashir", "Shoaib", "Ismail",
    "Tahir", "Javed", "Mehmood", "Hashim", "Umair", "Ansar", "Adil", "Raza", "Abrar"
]

# Shuffle names
random.shuffle(pakistani_boy_names)

# Function to assign grade
def get_grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
        return "A"
    elif percentage >= 70:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 50:
        return "D"
    else:
        return "F"

# Create student data
students = []
for roll_no, name in enumerate(pakistani_boy_names, start=1):
    maths = random.randint(33, 100)
    computer = random.randint(33, 100)
    physics = random.randint(33, 100)

    total = maths + computer + physics
    avg = round(total / 3, 2)
    percentage = round((total / 300) * 100, 2)  # 3 subjects, each out of 100
    grade = get_grade(percentage)

    student = {
        "Roll No": roll_no,
        "Name": name,
        "Maths": maths,
        "Computer": computer,
        "Physics": physics,
        "Total Marks": total,
        "Average Marks": avg,
        "Percentage": percentage,
        "Grade": grade
    }
    students.append(student)

# Save to CSV
with open("students_marksheet.csv", "w", newline="") as csvfile:
    fieldnames = ["Roll No", "Name", "Maths", "Computer", "Physics",
                  "Total Marks", "Average Marks", "Percentage", "Grade"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    writer.writerows(students)

print("✅ students_marksheet.csv file created successfully with totals, averages, percentage, and grade!")

```
